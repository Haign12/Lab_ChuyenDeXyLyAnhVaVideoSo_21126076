```python
yaml_content = """
path: /content/drive/MyDrive/Data_Mot17
train: images/train
val: images/train
nc: 1
names: ['person']
"""

with open("/content/drive/MyDrive/Data_Mot17/data.yaml", "w") as f:
    f.write(yaml_content.strip())
```

```python
from ultralytics import YOLO

model = YOLO("yolov8m.pt")

model.train(
    data="/content/drive/MyDrive/Data_Mot17/data.yaml",
    epochs=20,
    imgsz=512,
    batch=8,
    project="/content/drive/MyDrive/Trained_Model",
    name="Mot17",
    save = True,
    save_period = 1,
)
```
